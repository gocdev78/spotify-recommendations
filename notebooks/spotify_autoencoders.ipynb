```python
from torch.utils.data import Dataset
import torch


class SongDataset(Dataset):
    def __init__(self, df, include_metadata_columns = False):
        self.df = df
        self.include_metadata_columns = include_metadata_columns

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        features_numpy = row[feature_columns].to_numpy().astype(float)
        features_torch = torch.from_numpy(features_numpy).to(torch.float32)
        metadata = row[metadata_columns].to_numpy()
        
        if not self.include_metadata_columns:
            return features_torch

        return metadata, features_torch
```

```python
class CollateBatch:
    def __init__(self, batch):
        metadata, features = zip(*batch)
        self.features = torch.stack(features, dim=0)
        self.metadata = np.asarray(list(metadata), dtype=str)


    @staticmethod
    def build(batch):
        return CollateBatch(batch)
```

```python
from torch.utils.data import DataLoader

batch_size = 32

train_dataloader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_dataloader = DataLoader(test_dataset, batch_size=batch_size, shuffle=True, collate_fn=CollateBatch.build)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim


class SongAutoencoder(nn.Module):
    def __init__(self, input_dim, genre_vocab_size, genre_emb_dim, latent_dim):
        super(SongAutoencoder, self).__init__()
        self.return_latent = False
        self.genre_embedding = nn.Embedding(genre_vocab_size, genre_emb_dim)
        self.encoder = nn.Sequential(
            nn.Linear(input_dim + genre_emb_dim, latent_dim * 4),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(latent_dim * 4, latent_dim * 2),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(latent_dim * 2, latent_dim)
        )
        self.decoder = nn.Sequential(
            nn.Linear(latent_dim, latent_dim * 2),
            nn.ReLU(),
            nn.Linear(latent_dim * 2, latent_dim * 4),
            nn.ReLU(),
            nn.Linear(latent_dim * 4, input_dim + 1),
        )
    
    def forward(self, x):
        genre_id = x[:, -1].to(torch.int32)
        x = x[:, :-1]
        genre_emb = self.genre_embedding(genre_id)
        x = torch.cat([x, genre_emb], dim=1)
        latent = self.encoder(x)
        if self.return_latent:
            return latent
        reconstructed = self.decoder(latent)
        return reconstructed
```

```python
latent_space = 32
input_dim = len(df.columns) - 1 - len(metadata_columns) # all columns but genre_id which is embedded and metadata columns which are ignored
vocab_size = len(df['genre_id'].unique())
embed_dim = 128
```

```python
autoencoder = SongAutoencoder(input_dim, vocab_size, embed_dim, latent_space)
```

```python
autoencoder(next(iter(train_dataloader))).shape
```

```python
train_dataloader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_dataloader = DataLoader(test_dataset, batch_size=batch_size, shuffle=True, collate_fn=CollateBatch.build)
```

```python
epochs = 30
criterion = nn.MSELoss()
optimizer = optim.Adam(autoencoder.parameters(), lr=0.001, weight_decay=1e-5)
best_val_loss = float("inf")
best_model_path = "best_autoencoder.pth"
```

```python
for epoch in range(epochs):
    autoencoder.train()
    epoch_loss = 0.0
    
    for data in train_dataloader:
        optimizer.zero_grad()
        reconstructed = autoencoder(data)
        loss = criterion(reconstructed, data)
        loss.backward()
        optimizer.step()
        epoch_loss += loss.item()
    
    avg_train_loss = epoch_loss / len(train_dataloader)

    autoencoder.eval()
    val_loss = 0.0
    with torch.no_grad():
        for collate_batch in test_dataloader:
            reconstructed = autoencoder(collate_batch.features)
            loss = criterion(reconstructed, collate_batch.features)
            val_loss += loss.item()
    
    avg_val_loss = val_loss / len(test_dataloader)

    if avg_val_loss < best_val_loss:
        best_val_loss = avg_val_loss
        torch.save(autoencoder.state_dict(), best_model_path)
        print(f"✅ Model saved at epoch {epoch+1} with val_loss: {avg_val_loss:.6f}")

    print(f"Epoch {epoch+1}/{epochs}, Train Loss: {avg_train_loss:.6f}, Val Loss: {avg_val_loss:.6f}")

print("Training complete! Best model saved as", best_model_path)
```

```python
autoencoder_eval = SongAutoencoder(input_dim, vocab_size, embed_dim, latent_space)
autoencoder_eval.load_state_dict(torch.load(best_model_path, weights_only=True))
autoencoder_eval.eval()
autoencoder_eval.return_latent = True
```

```python
preds = {
    "latent_vectors": [],
    "song": [],
    "album": [],
    "id": [],
    "artist": [],
}

with torch.no_grad():
    for collate_batch in test_dataloader:
        latent_vector = autoencoder(collate_batch.features)
        metdata = collate_batch.metadata
        preds["latent_vectors"].extend(latent_vector.numpy())
        preds["id"].extend(collate_batch.metadata[:, 0])
        preds["song"].extend(collate_batch.metadata[:, 1])
        preds["album"].extend(collate_batch.metadata[:, 2])
        preds["artist"].extend(collate_batch.metadata[:, 3])
```
